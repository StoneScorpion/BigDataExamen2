```python
import time
import pymongo 
import requests

from bs4 import BeautifulSoup
from selenium import webdriver
from urllib.request import urlopen

from pymongo import MongoClient
```

```python
SITE_URL = 'https://alianzaverde.org.co'
```

```python
DRIVER_PATH = './geckodriver.exe'
```

```python
browser = webdriver.Firefox(executable_path = DRIVER_PATH)
```

```python
def make_request(browser, relative_path):
    browser.get(SITE_URL + relative_path)
    time.sleep(5)
    check_height = browser.execute_script('return document.body.scrollHeight;')
    while True:
        time.sleep(5)
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(5)
        height = browser.execute_script('return document.body.scrollHeight;')
        if height == check_height: 
            break 
        check_height = height
        
    return BeautifulSoup(browser.page_source, 'html.parser')
```

```python
soup = make_request(browser, '/lideres-verdes/congreso-2022/senado-2022')
```

```python
layout = soup.find('section',class_ = "main-body")
```

```python
print(layout)
```

```python
blocks = layout.find_all(class_ = 'sppb-col-md-3', recursive = True)
print(len(blocks))
```

```python
for block in blocks:
    try:
        photo = block.find('img', class_ = 'sppb-img-responsive').get('src')
        print('Photo: ' + photo)
    except FileNotFoundError:
        print('No hay foto')
    
    try:
        description = block.find('span', class_ = 'sppb-person-designation').get_text()
        print('Description: ' + description)
    except FileNotFoundError:
        print('No hay descripcion')
    
    try:
        for href in block.find_all('a',href=True):
            print('Social Media: ' + href.get('href'))
        print()
    except FileNotFoundError:
        print('No tiene redes sociales')
```
